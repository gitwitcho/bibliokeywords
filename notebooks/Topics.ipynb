```python
import sys
from pathlib import Path

# Add the src directory to the Python path
src_path = Path("../") / "src"
if src_path.resolve() not in sys.path:
    sys.path.insert(0, str(src_path.resolve()))

from config import *
from utilities import *
from topics import *
```

```python
# Input parameters
# -----------------------
biblio_project_dir = 'example_project'              # directory for the data and models of your bibliographic project
biblio_input_dir = 'processed'                      # directory containing the input file
biblio_input_file = 'biblio_example_all.csv'        # input file (bibligraphic dataset)

output_dir = 'results'                              # directory where you want to save the bibliographic dataset with the topic information
output_file = f'topics_abstract_example_500.csv'    # filename of the bibliographic dataset; leave empty if you don't want to save the data

n_rows = 500                # the maximum number of rows read for each dataset; set to '0' if you want to read all the data

topic_col = 'abstract'      # the column containing the text on which the BERT topic model is run
sampling = False            # will randomly sample n_rows from the input bibliographic dataset
# -----------------------
```

```python
# 1. Read the bibliographic datasets
biblio_df = read_biblio_csv_files_to_df(biblio_project_dir = biblio_project_dir, 
                                        input_dir = biblio_input_dir,
                                        input_files = biblio_input_file,
                                        biblio_source = BiblioSource.BIBLIO,
                                        n_rows = n_rows,
                                        sample = sampling)

# 2. Generate topics with BERTopic
model, merged_df = generate_bert_topics(biblio_df_ = biblio_df,
                                        input_col = topic_col,
                                        verbose = True)

# 3. Save the merged biblio_df/topics to a CSV file
if output_file:
    write_df(biblio_df = merged_df,
            biblio_project_dir = biblio_project_dir,
            output_dir = output_dir,
            output_file = output_file)
```
